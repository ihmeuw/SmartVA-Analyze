```python
def load_results(dname):
    """Load all results from a run of SmartVA-Analyze
    
    Parameters
    ----------
    dname : str, path to directory where results are stored
    
    Results
    -------
    return results_dict, a dict of pd.DataFrames, including
    data for '{module}-csmf.csv', '{module}-prediction.csv',
    """
    import glob
    results_dict = {}
    csv_files = glob.glob(dname + '/*.csv') # python list of files matching this regex
    csv_files += glob.glob(dname + '/intermediate-files/*.csv')
    
    for fname in csv_files: # FIXME: currently unix only
        key = fname.split('/')[-1] # find fname
        key = key.replace('.csv', '')
        results_dict[key] = pd.read_csv(fname)
        
    
    return results_dict

results = load_results(dname)
results.keys()
```

```python
import smartva.data.common_data

def free_text(results, module, i):
    txt = 'Free Text:\n'
    
    s = results['cleanheaders'].loc[i]  # FIXME: need to deal with module somehow
    
    for col in smartva.data.common_data.FREE_TEXT_VARS:
        txt_i = str(s.get(col, ''))
        if txt_i and txt_i != 'nan':
            txt += ' - ' + col + ': ' + txt_i + '\n'
            
    if txt == 'Free Text:\n':
        txt += ' (none)'

    return txt

print free_text(results, module, 0)
```

```python
def merge_tariff_vals(results, module, i, df, J, additional_causes):
    """Add j columns to pd.DataFrame df based on results
    for VAI i
    
    Parameters
    ----------
    results : dict of pd.DataFrames
    i : int, which VAI to show
    df : pd.DataFrame to add columns to
    J : int, number of top causes to include
    additional_causes : list of ints, causes to include in addition to top J
    """
    cause_ranks = results[module+'-tariff-ranks'].loc[i]
    cause_ranks = cause_ranks.drop(['sid'])
    causes_to_merge = cause_ranks.sort_values().iloc[:J]
    causes_to_merge = list(causes_to_merge.index.astype(int))
    
    # merge in additional causes
    for cause_num in additional_causes:
        if cause_num not in causes_to_merge:
            causes_to_merge.append(cause_num)
    
    t = []
    for cause_num in causes_to_merge:
        cause_num = int(cause_num)
        col = cause_num
        t.append({' Cause Num (46)': col, 'Cause Name':cause_map[module][cause_num]})
        df[col] = tariff_matrix[module].iloc[cause_num-1]
        
        # round to nearest .5
        df[col] = .5 * np.round(df[col].astype(float)*2)
        
    display(pd.DataFrame(t))
        
    # add additional rows for sum, sum after exclusions, rank, and rank after exclusions
    s = {}
    for col in df.columns:
        if col in ['symptom', 'question']:
            s[col] = ''
        elif col == 'answer':
            s[col] = 'Total Score'
        else:
            s[col] = df[col].sum()
    
    df = df.append(s, ignore_index=True)

    # replace 0 with blank in tariff table
    for cause_num in causes_to_merge:
        cause_num = int(cause_num)
        col = cause_num
        df[col] = np.where(df[col] == 0, '', df[col])

    for row in ['-tariff-scores', '-external-ranks', '-tariff-ranks']:
        s = {}
        for col in df.columns:
            if col in ['symptom', 'question']:
                s[col] = ''
            elif col == 'answer':
                s[col] = row
            else:
                s[col] = results[module+row].loc[i,str(col)]

        df = df.append(s, ignore_index=True)

    return df
```

```python
from IPython.display import display

def show_tariffs(results, i):
    """Show deep-dive results on a single tariff assignment as Jon J developed:
      symptoms, tariffs, sum, and ranks for an individual VAI
    
    Parameters
    ----------
    results : dict, as returned by load_results
    i : int, which VAI to show
    TODO: module, filtering by cause
    
    Results
    -------
    displays pd.DataFrames on single VAI
    """
    
    module = 'child'
    df = results[module + '-predictions'].loc[[i]]
    
    display(df)
    
    # display free text
    print free_text(results, module, i)
    
    df = results[module + '-symptom'].loc[[i]]
    #df = df.filter(regex='s[0-9]+') # keep only symptom columns
    
    # transpose and keep only non-zero symptoms
    df = df.T
    df.columns = ['answer']
    df = df[(df.answer==1)|(df.index.isin(['age', 'sex']))]  # all non-zero symptoms and also include age and sex columns
    
    # merge in question corresponding to each symptom
    df['question'] = pd.Series(df.index, index=df.index).map(symptom_map[module].gc13_label)
    # shorten questions
    for extra_txt in ['During the illness that led to death, ',
                      'Was \[name\] ', 'did \[name\] ',
                      "Did \[name\]'s "]:
        df.question = df.question.str.replace(extra_txt, '')
        
    # get symptoms into a column, from index
    df['symptom'] = df.index
    
    # re-order columns for prettiness
    df = df.dropna().filter(['symptom', 'question', 'answer'])
    
    # merge in tariff values for causes with lowest tariff ranks
    J = 3 # number of causes to merge
    additional_causes = [9, 10] # malaria and measles
    df = merge_tariff_vals(results, module, i, df, J, additional_causes)
    
    display(df)
    
show_tariffs(results, 0)
```
